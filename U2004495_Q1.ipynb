```
"""
author: Mohamad Zuhaily Bin Mahmud
matrics num. : U2004495
"""
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

##############################################################################
#Question 1a
##############################################################################

def slider(B):

  #setting up data and figure
  fig = plt.figure()
  ax1=fig.add_subplot()
  c = 299792458 #light speed
  gamma=1/np.sqrt(1-(B**2))  #lorentz factor
  x=np.arange(0,10,0.1) #original x value
  t=np.arange(0,10,0.1) #original ct value

  #calculation
  alpha = np.arctan(B)
  k=x*np.cos(alpha)
  l=t*np.sin(alpha)

  #plot
  ax1.plot(k,l,label='Space, x`',color='red')
  ax1.plot(l,k,label='Time, ct`',color='blue')
  fig.legend(loc='upper right')
  ax1.title.set_text('Question 1\nSpace-time transformation\ndue to Lorentz transformation') 
  ax1.set_xlabel('Space, x`')
  ax1.set_ylabel('Time, ct`')
  if B==1:
    print('Error! B ≠ 1.0')

  plt.show()

interact(slider,B =(0,1,0.01))

##############################################################################
#Question 1b
##############################################################################

def zlider_2():
  fig = plt.figure()
  ax2=fig.add_subplot()

  #create value
  B=np.arange(0,1,0.01) # beta value
  alpha = np.arctan(B) #angle value

  #plot
  ax2.plot(B,alpha,color='red')
  ax2.title.set_text('\n\nQuestion 1b\narctan B vs B') 
  ax2.set_xlabel('B')
  ax2.set_ylabel('arctan B, a (rad)')
  plt.show()
  
zlider_2()

```
